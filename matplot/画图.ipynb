```python



def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2.- 0.2, 1.03*height, '%s' % int(height))
        

name_list = ['1y', '2y','1c+1r']
num_list = [2598.947253, 506.824365, 6346.665171]
num_list2 = [259.692398786544, 375.684004306793,2769.4835484027863]
x = list(range(len(num_list)))
total_width, n = 0.8, 2
width = total_width / n
z1 = plt.bar(x, num_list, width=width, label='mnist', fc='b')
for i in range(len(x)):
    x[i] += width
z2 = plt.bar(x, num_list2, width=width, label='cifar10', tick_label=name_list, fc='g')



autolabel(z1)
autolabel(z2)

plt.legend()
plt.title("卷积网络训练耗时对比")
# for x,y in zip(x,y):
#     plt.text(x+0.02,y+1,y,ha='center',va='bottom')
plt.xlabel('网络 （层）') #X轴标签
plt.ylabel("耗时 （秒）",va='bottom') #Y轴标签

plt.savefig('./train_cnn.jpg')   #图片的存储
plt.show()


plt.close()   #关闭matplotlib
```

```python
from matplotlib import pyplot
import matplotlib.pyplot as plt
 
# names = range(0,13)
# names = [str(x) for x in list(names)]
names = ['1','2','4','8','10']

 
x = range(len(names))
y_mnist = [0.9164,0.9894,0.9901,0.987,0.9841]
y_cifar  = [0.3304,0.3557,0.3567,0.3189,0.2966]
#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
#pl.xlim(-1, 11)  # 限定横轴的范围
#pl.ylim(-1, 110)  # 限定纵轴的范围
 
 
plt.plot(x, y_mnist, marker='o', mec='r', mfc='w',label='mnist_acc')
plt.plot(x, y_cifar, marker='*', ms=10,label='cifar_acc')
plt.legend()  # 让图例生效

plt.title("卷积网络准确率对比")
plt.xticks(x, names, rotation=1)
 
plt.margins(0)
plt.subplots_adjust(bottom=0.10)
plt.xlabel('循环网络 （层）') #X轴标签
plt.ylabel("准确率") #Y轴标签
pyplot.yticks([0.00,0.500,1.00,1.500])
#plt.title("A simple plot") #标题
# plt.savefig('D:\\f1.jpg',dpi = 900)
plt.savefig('rnn_acc.jpg')

plt.show()
```
